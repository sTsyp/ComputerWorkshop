```python
#импорт библиотек
import numba
import numpy as np
import holoviews as hv
hv.extension('bokeh')
from holoviews import opts
from holoviews.streams import RangeXY
from numba import jit
```

**1. Реализовать функции и откомпилировать их при помощи модуля `Numba`**

а) расчет $\lambda$ для точки на комплексной плоскости

```python
renderer = hv.renderer('bokeh')
@numba.jit(nopython=True)
def c_l(x, y, m_i):
    c = complex(x, y)
    z = 0.0j
    for i in range(m_i):
        z = z ** 3 + c
        if abs(z) >= 2.0:
            return i
    return m_i
```

- x, y - координаты точки на комплексной плоскости.
- m_i - максимальное количество итераций.
- Функция итеративно вычисляет значение z с использованием формулы  и проверяет, выполняется ли условие выхода (что модуль z становится больше или равен 2.0). Если условие выполняется, функция возвращает количество итераций, на котором потребовалось достичь этого условия. В противном случае, если максимальное количество итераций m_i достигнуто, функция возвращает m_i.

б) расчет $\lambda$ на регулярной решетке на комплексной плоскости (эта функция должна быть распараллелена при помощи `prange`)

```python
@numba.jit(nopython=True, parallel=True)
def c_l_g(min_x, max_x, min_y, max_y, img, i):
    h = 600
    w = 600
    size_x = (max_x - min_x) / w
    size_y = (max_y - min_y) / h
    for x in numba.prange(w):
        re = min_x + x * size_x
        for y in numba.prange(h):
            im = min_y + y * size_y
            color = c_l(re, im, i)
            img[y, x] = color
    return img
```

- min_x, max_x, min_y, max_y - границы области в комплексной плоскости, на которой вычисляется множество Мандельброта.
- img - массив NumPy, в который будут записаны результаты вычислений (множество Мандельброта).
- iters - максимальное количество итераций для вычисления каждой точки.
- Функция начинает с задания размеров изображения (h и w) и вычисляет размеры пикселя size_x и size_y на основе границ и размеров изображения. Затем она обходит регулярную сетку в комплексной плоскости (по x и y) и вызывает функцию cl для каждой точки с координатами (real, imag), где real и imag - это реальная и мнимая части текущей точки. Результат (color) записывается в массив img в соответствии с координатами (y, x).

- numba.prange используется для распараллеливания циклов



**2. Cоздать интерактивную карту при помощи класса `DynamicMap` модуля `holoviews`**

```python
def get_frac(x_range, y_range):
    x0, x1 = x_range 
    y0, y1 = y_range
    img = np.zeros((600, 600), dtype=np.uint8)
    arr = c_l_g(x0, x1, -y1, -y0, img, 200)
    return hv.Image(arr, bounds=(x0, y0, x1, y1))

r_s = RangeXY(x_range=(-1.25, 1.25), y_range=(-1.25, 1.25))

dynamic_map = hv.DynamicMap(get_frac, streams=[r_s]).hist(log=True)
dynamic_map.opts(
    opts.Histogram(framewise=True, logy=True, width=200),
    opts.Image(cmap='kr', logz=True, height=600, width=600, xaxis=None, yaxis=None)
)
```

- x_range, y_range: Границы области на комплексной плоскости.
- image: Массив NumPy для хранения изображения множества Мандельброта.
- iters: Максимальное количество итераций для вычисления каждой точки.
- r_s: Объект RangeXY, определяющий начальную область для отображения.
- get_frac(x_range, y_range): Создает изображение множества Мандельброта на заданной области комплексной плоскости.
- c_l_g(min_x, max_x, min_y, max_y, image, iters): Расчитывает значения пикселей множества Мандельброта на регулярной решетке.
